# Interoperability with Matplotlib
Scientists, obviously, need assistance from their trusted machines to generate clear visuals from *piles of messy nonsense*. When plain tables from **pandas** is just not enough, **Matplotlib** can be used in conjunction with it to provide graphs for these abstract data.

As usual, don't forget to import needed libraries!

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Firstly we import our data from a CSV.
Also as usual, we examine it for a bit before doing anything else.

```python
air_quality = pd.read_csv("air_quality_no2.csv", index_col=0, parse_dates=True)
air_quality.head()
```

You can let it work a default one out, or choose some columns to include.

```python
air_quality.plot()
air_quality["station_paris"].plot()
```

You can, of course, chose a specific type of visual that fits your need for *science(TM)*.
Here we choose to visually compare the $NO_2$ values in London and Paris.

```python
air_quality.plot.scatter(x="station_london", y="station_paris", alpha=0.5)
```

You don't actually rely on PyCharm's auto-completing abilities to see how your data can be plotted.

```python
[
    method_name
    for method_name in dir(air_quality.plot)
    if not method_name.startswith("_")
]
```

```python
# Fancy box plotting!
air_quality.plot.box()
```

You can separate each column out too, or, customize it even further!
And remember that you can always save down your plot and share it with others.

```python
fig, axs = plt.subplots(figsize=(12, 4))
air_quality.plot.area(ax=axs)
axs.set_ylabel("NO$_2$ concentration")
fig.savefig("no2_concentrations.png")
```

Plots created by pandas is Matplotlib objects. Any operation applicable on it is available here! You may want to look into that library some day.
