# Introduction to pandas
pandas is a powerful, flexible and easy to use open source data analysis and manipulation tool in Python.

And of course, you want to import **pandas** as **pd** to make our lives easier.

```python
import pandas as pd
print("It's pandas, version", pd.__version__, "!")
```

## DataFrame and Series
**DataFrame** in pandas represents what is so-called *spreadsheets*. It's usually similar to what is called *table* in relational databases or **data.frame** in R. Every column has their associated name.

```python
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)

df
```

Each column in a **DataFrame** is a **Series**, a sequence of data that (maybe) has its own name.

*HINT* - **DataFrames** in pandas work like *dictionaries* in python, which can be indexed with the key of each entry.

Also, a series can be manually created.

```python
df["Age"]

ages = pd.Series([22, 35, 58], name="Age")
ages
```

## Statistical computing with DataFrame and Series
We can, of course, do *scientist* stuff with these two data structures. For example, we can get the maximum of certain value of let pandas **describe** some sort of data in a *human-friendly* way without needing you to traverse through millions lines of seemingly-same numbers.

```python
# Maximum of a specific column in a DataFrame
df["Age"].max()
# Maximum of a Series
ages.max()
```

```python
# Analyse the numeric data of it for me!
df.describe()
```
